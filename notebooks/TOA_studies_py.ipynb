```python
import ROOT
```

```python
#ROOT.enableJSVis()
ROOT.disableJSVis()
```

```python
c = ROOT.TCanvas("myCanvasName","The Canvas Title",800,600)
```

```python
file = ROOT.TFile.Open("../data/hexa1/RUN_290317_1109_new.root")
```

```python
sk2cms = file.Get("sk2cms")
```

```python
sk2cms.Draw("toa_rise[]:Iteration$","","colz"); c.Draw()
```

```python
sk2cms.Draw("toa_rise[]:Iteration$","toa_rise[] > 4 && toa_rise[] < 3000","colz"); c.Draw()
```

```python
c2 = ROOT.TCanvas("c2","The Canvas Title",800,600)
c2.Divide(2,2)
```

```python
for chip in range(4):
    c2.cd(chip+1)
    var = "toa_rise[%i][]" % chip
    draw_line = var + ":Iteration$ >> chip_%i" % (chip)
    cut_line = ""
    sk2cms.Draw(draw_line,cut_line,"colz")
c2.Draw()
```

```python
for chip in range(4):
    c2.cd(chip+1)
    var = "toa_rise[%i][]" % chip
    draw_line = var + ":Iteration$ >> zoom_chip_%i" % (chip)
    cut_line = var + " > 4" + "&&" + var + " < 3000"
    sk2cms.Draw(draw_line,cut_line,"colz")
c2.Draw()
```

```python
c3 = ROOT.TCanvas("c3","The Canvas Title",800,600)
c3.Divide(2,2)

for chip in range(4):
    c3.cd(chip+1)
    var = "tot_fast[%i][]" % chip
    draw_line = var + ":Iteration$ >> tot_chip_%i" % (chip)
    cut_line = "" #var + " > -4" + "&&" + var + " < 30000"
    sk2cms.Draw(draw_line,cut_line,"colz")
c3.Draw()
```

```python
c3 = ROOT.TCanvas("c3","The Canvas Title",800,600)
c3.Divide(2,2)

for chip in range(4):
    c3.cd(chip+1)
    var = "hg[%i][0][]" % chip
    draw_line = var + ":Iteration$ >> hg_chip_%i" % (chip)
    cut_line = "roll[0] == 0" #var + " > -4" + "&&" + var + " < 30000"
    sk2cms.Draw(draw_line,cut_line,"colz")
c3.Draw()
```
